```python
#leer archivo csv
#imprimir 50 lineas
file=open(r'./Height of Male and Female by Country 2022.csv')
csvreader=csv.reader(file)
header=[]
header=next(csvreader)
print(header)

rows=[]
for row in csvreader:
    rows.append(row)

print(rows[:50])

```

```python
#Calcular la media, mínima y máxima altura de hombres y mujeres del data set completo
from statistics import mean
import numpy as np

BD=np.array(rows)
hombres=BD[0:,2]


mujeres=BD[0:,3]

alt_mujeres =[float(alt) for alt in mujeres]
alt_hombres =[float(alt) for alt in hombres]


print("La media de altura de las mujeres es: {:.3f}".format(mean(alt_mujeres)))
print("La media de altura de las mujeres es: {:.3f}".format(mean(alt_hombres)))
print("La altura mínima de las mujeres es: {:.3f}".format(min(alt_mujeres)))
print("La altura maxima de las mujeres es: {:.3f}".format(max(alt_mujeres)))
print("La altura mínima de los hombres es: {:.3f}".format(min(alt_hombres)))
print("La altura maxima de las hombres es: {:.3f}".format(max(alt_hombres)))
```
